```python
import pandas as pd
import joblib
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.dummy import DummyRegressor

# 1. Load Data
train = pd.read_csv('/kaggle/input/spl-utspan-data-challenge-2026/train.csv')
test = pd.read_csv('/kaggle/input/spl-utspan-data-challenge-2026/test.csv')

# Load the provided target scalers
scaler_angle = joblib.load('/kaggle/input/spl-utspan-data-challenge-2026/scaler_angle.pkl')
scaler_depth = joblib.load('/kaggle/input/spl-utspan-data-challenge-2026/scaler_depth.pkl')
scaler_lr = joblib.load('/kaggle/input/spl-utspan-data-challenge-2026/scaler_left_right.pkl')

# Define target columns
targets = ['angle', 'depth', 'left_right']
X = train.drop(columns=['id', 'shot_id', 'participant_id'] + targets)
y = train[targets]

# 2. Setup Feature Preprocessing 
numeric_features = X.select_dtypes(include=['float64', 'int64']).columns
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline(steps=[
            ('imputer', SimpleImputer(strategy='median')),
            ('scaler', StandardScaler())
        ]), numeric_features)
    ])

# 3. Dummy System
# We will predict the mean of the training targets
model = DummyRegressor(strategy='mean')

# Fit preprocessing on features
X_processed = preprocessor.fit_transform(X)
model.fit(X_processed, y)

# 4. Predict and Apply Target Scalers
X_test_processed = preprocessor.transform(test.drop(columns=['id', 'shot_id', 'participant_id'], errors='ignore'))
raw_preds = model.predict(X_test_processed)

# Re-scale the predictions using your .pkl files
# Note: Usually, if the targets are 'scaled_angle', you use transform(). 
# If they are 'angle' and you need them scaled for submission, use transform().
scaled_preds_angle = scaler_angle.transform(raw_preds[:, 0].reshape(-1, 1))
scaled_preds_depth = scaler_depth.transform(raw_preds[:, 1].reshape(-1, 1))
scaled_preds_lr = scaler_lr.transform(raw_preds[:, 2].reshape(-1, 1))
```

```python
# Create submission dataframe
submission = pd.DataFrame({
    'id': test['id'],
    'scaled_angle': scaled_preds_angle.flatten(),
    'scaled_depth': scaled_preds_depth.flatten(),
    'scaled_left_right': scaled_preds_lr.flatten()
})

# Export
submission.to_csv('dummy_submission.csv', index=False)
print("Submission with applied target scalers is ready!")
```
